# 模型微调

```python
# 相对小参数量的模型，我们只需要使用 1 张卡（第0张）
# 如果你有多张卡，使用 "," 分隔显卡 ID即可，如: 0,1,2,3 （4卡）
!export CUDA_VISIBLE_DEVICES=0

!swift sft --model_id_or_path ./LLM-Research/Meta-Llama-3___1-8B-Instruct \
    --model_type llama3_1-8b-instruct \
    --sft_type lora \
    --tuner_backend peft \
    --template_type AUTO \
    --dtype AUTO \
    --output_dir output \
    --dataset ./ruozhiba-llama3/ruozhiba_qa.json \
    --train_dataset_sample -1 \
    --num_train_epochs 5 \
    --max_length 2048 \
    --check_dataset_strategy warning \
    --lora_rank 8 \
    --lora_alpha 32 \
    --lora_dropout_p 0.05 \
    --lora_target_modules ALL \
    --gradient_checkpointing true \
    --batch_size 1 \
    --weight_decay 0.1 \
    --learning_rate 1e-4 \
    --gradient_accumulation_steps 16 \
    --max_grad_norm 0.5 \
    --warmup_ratio 0.03 \
    --eval_steps 100 \
    --save_steps 100 \
    --save_total_limit 2 \
    --logging_steps 10 
```

```python
!export CUDA_VISIBLE_DEVICES=0
# ！替换为你刚刚训练结果中的模型权重文件！
!swift infer \
    --ckpt_dir "/mnt/workspace/output/llama3_1-8b-instruct/v1-20240801-215508/checkpoint-92" \
    --load_dataset_config true \
    --max_new_tokens 2048 \
    --temperature 0.1 \
    --top_p 0.7 \
    --repetition_penalty 1. \
    --do_sample true \
    --merge_lora false 
```

```python
!export CUDA_VISIBLE_DEVICES=0
# 测试原始模型
!swift infer \
    --model_type llama3_1-8b-instruct \
    --ckpt_dir "" \
    --load_args_from_ckpt_dir true \
    --model_id_or_path ./LLM-Research/Meta-Llama-3___1-8B-Instruct \
    --load_dataset_config true \
    --temperature 0.1 \
    --top_p 0.7 \
    --repetition_penalty 1. \
    --do_sample true \
    --dataset ./ruozhiba-llama3/ruozhiba_qa.json
```

```python
# AWQ
# 使用 alpaca-zh alpaca-en sharegpt-gpt4:default 作为量化数据集
#!export CUDA_VISIBLE_DEVICES=0
#!export NPROC_PER_NODE=1
#!swift export \
#    --model_type llama3_1-8b-instruct \
#    --model_id_or_path ./LLM-Research/Meta-Llama-3___1-8B-Instruct \
#    --quant_device_map="cuda:0" \
#    --quant_bits 4 \
#    --dataset alpaca-zh alpaca-en sharegpt-gpt4:default --quant_method awq

# GPTQ
# 使用`alpaca-zh alpaca-en sharegpt-gpt4:default`作为量化数据集
# gptq量化请先查看此issue: https://github.com/AutoGPTQ/AutoGPTQ/issues/439
#!export OMP_NUM_THREADS=14
#!export CUDA_VISIBLE_DEVICES=0
#!swift export \
#    --model_type llama3_1-8b-instruct \
#    --quant_bits 4 \
#    --dataset alpaca-zh alpaca-en sharegpt-gpt4:default --quant_method gptq

# BNB
!export CUDA_VISIBLE_DEVICES=0
!swift infer \
    --quantization_bit 4 \
    --quant_method bnb \
    --model_type llama3_1-8b-instruct \
    --ckpt_dir "" \
    --load_args_from_ckpt_dir true \
    --model_id_or_path ./LLM-Research/Meta-Llama-3___1-8B-Instruct \
    --load_dataset_config true \
    --temperature 0.1 \
    --top_p 0.7 \
    --repetition_penalty 1. \
    --do_sample true \
    --dataset ./ruozhiba-llama3/ruozhiba_qa.json
```
